```python
def load_data_old(file_name):
    data = open(file_name).read()
    sentences = data.split("\n\n")
    X = []
    Y = []
    for sentence in sentences:
        lines = sentence.split('\n')
        token_list = []
        label_list = []
        for line in lines:
            split_line = line.split('\t')
            if len(split_line) != 2:
                continue
            word, ner_tag = split_line
            token_list.append(word)
            if ner_tag[0] == 'I' and (len(label_list) == 0 or label_list[-1][0] == 'O'):
                label_list.append('B' + ner_tag[1:])
            else:
                label_list.append(ner_tag)
        if len(label_list) > 1:
            X.append(token_list)
            Y.append(label_list)
    return (X, Y)
```

```python
def load_data(file_name):
    data = open(file_name).read()
    sentences = data.split("\n\n")
    X = []
    Y = []
    for sentence in sentences:
        lines = sentence.split('\n')
        token_list = []
        label_list = []
        for line in lines:
            split_line = line.split('\t')
            if len(split_line) != 2:
                token_list.append([])
                label_list.append([])
                continue
            word, ner_tag = split_line
            token_list.append(word)
            if ner_tag[0] == 'I' and (len(label_list) == 0 or label_list[-1][0] == 'O'):
                label_list.append('B' + ner_tag[1:])
            else:
                label_list.append(ner_tag)
        X.append(token_list)
        Y.append(label_list)
    X = X[:-1]
    Y = Y[:-1]
    return (X,Y)
```

```python
def word_shape(i, tokens):
    word = ''
    for char in tokens[i]:
      if char.isalpha():
        word+= 'A'
      elif char.isdigit():
        word+= 'n'
      elif char in "./:%-":
        word+= 'P'
      else:
        word+= '-'
    return {'word shape': word}
```

```python
def featurize_one_sentence(tokens, feature_function_list):
    feature_dict_seq = []
    for i in range(len(tokens)): # ทุก token
        feature_dict = {}
        for feature_fn in feature_function_list: # ทุก feature function
            feature_dict.update(feature_fn(i, tokens))
        feature_dict_seq.append(feature_dict)
    return feature_dict_seq
    
def t_a_e(X_train, feature_function_list):
    features = [featurize_one_sentence(x, feature_function_list) for x in X_train]
    return features   

crf = sklearn_crfsuite.CRF(algorithm='l2sgd', all_possible_states=True, max_iterations=100) # all_possible_states=True
crf.fit(t_a_e(Xtrain, ff_list_exp), Ytrain)
pred = crf.predict([featurize_one_sentence(x, ff_list_exp) for x in dev])
#print(sklearn_crfsuite.metrics.flat_classification_report(Ydev, pred))
#print(sklearn_crfsuite.metrics.flat_f1_score(Ydev, pred, average='macro'))
```
