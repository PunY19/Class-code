```python
f = open("verb.csv")
dict_verb = {}
for verbs in f:
  verb = verbs.strip().split(",")
  dict_verb[verb[0]] = verb
del dict_verb['"be    (am']
del dict_verb['']
dict_verb["be"] = ["is", "am", "are", "was", "were", "been", "being", "be", " "]
dict_verb
```

```python
 def score(sentence, tokenizer, model):
    tokenize_input = tokenizer.encode(sentence)
    tensor_input = torch.tensor([tokenize_input])
    loss=model(tensor_input, labels=tensor_input)[0]
    return -loss.detach().numpy()
```

```python
def finalmaikatry(original):
    det = ['', 'a', 'an', 'the']
    prep = ['', 'about', 'at', 'by', 'for', 'from', 'in', 'of', 'on', 'to', 'with']
    answer = []
    for sentences in original:
        sentence = nlp(sentences.strip())
        word = [str(i) for i in sentence]
        if word[0][0].islower():
            word[0][0].upper()
        oriscore = score(sentences, tokenizer, model) 
        #print(oriscore)
        #print(sentences)
        for i in range(len(word)):
            if word[i].lower() in det:
                origin = score(' '.join(word), tokenizer, model)
                #ans = sentences
                for num in range(len(det)):
                    candidate = ' '.join(word[:i] + [det[num]] + word[i+1:])
                    newscore = score(candidate, tokenizer, model)
                    if  newscore > origin:
                        origin = newscore
                        #ans = candidate
                        if i == 0:
                          word[i] = det[num].capitalize()
                        else:
                          word[i] = det[num]
                    else:
                        continue
            elif word[i] in prep:
                origin = score(' '.join(word), tokenizer, model)
                #ans = sentences
                for num in range(len(prep)):
                    candidate = ' '.join(word[:i] + [prep[num]] + word[i+1:])
                    newscore = score(candidate, tokenizer, model)
                    if  newscore > origin:
                        origin = newscore
                        word[i] = prep[num]
                        #ans = candidate
                        #if i == 0:
                          #word[i] = prep[num].capitalize()
                        #else:
                        
                    else:
                        continue
            elif sentence[i].pos_ == 'VERB' or sentence[i].pos_ == "AUX":
                origin = score(' '.join(word), tokenizer, model)
                ans = str(sentence[i].lemma_)
                if ans in dict_verb:
                      for w in dict_verb[ans]:
                          candidate = ' '.join(word[:i] + [w] + word[i+1:])
                          newscore = score(candidate, tokenizer, model)
                          if  newscore > origin:
                              origin = newscore
                              #ans = candidate
                              word[i] = w
                          else:
                              continue                          
            else:
              #ans = sentences
              newscore = score(' '.join(word), tokenizer, model)
        #word[0] = word[0].capitalize()
        ans = ' '.join(word).strip()
        ans = ans[0].upper() + ans[1:]
        #print(newscore, ans)
        answer.append([oriscore, newscore, ans])
    return answer
```
